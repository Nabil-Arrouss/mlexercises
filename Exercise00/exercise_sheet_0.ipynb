```python
path = 'Data.csv'

dataset = pd.read_csv(path)

dataset.info()
```

```python
print(f'Number of rows and columns in the dataset: {dataset.shape}')
```

```python
# Separate the features (X) and the target (y)
# Features 
X = dataset.drop('price', axis=1).values  
# Target 
y = dataset['price'].values  

# the dimensions of X, W, and B
# X is the feature matrix (N samples, d features)
N, d = X.shape
print(f"Feature tensor X dimensions: {X.shape}")

W = np.random.randn(d)  # Weights for each feature
B = np.random.randn()  # Bias term

print(f"Weight tensor W dimensions: {W.shape}")
print(f"Bias tensor B dimensions: {B}")

# Affine Transformation:  X' = X · W + B 
def affine_transformation(X, W, B):
    return np.dot(X, W) + B

X_prime = affine_transformation(X, W, B)
print(f"Affine transformation X' dimensions: {X_prime.shape}")

# Sigmoid Function: σ(X') = 1 / (1 + e^(-X'))
def sigmoid(X_prime):
    return 1 / (1 + np.exp(-X_prime))

sigma_output = sigmoid(X_prime)
print(f"Sigmoid output (σ(X')) dimensions: {sigma_output.shape}")
```

```python
def loss_function(X, y_hat, W, B):
    """Calculates the Mean Squared Error (MSE) loss"""
    y_pred = sigmoid(affine_transformation(X, W, B))
    squared_error = np.sum((y_hat - y_pred)**2) 
    return squared_error

# --- Random Search Implementation ---
N_ITERATIONS = 50000 
best_loss = float('inf')
best_W = None
best_B = None


print(f"Starting Random Search with {N_ITERATIONS} iterations (D={d} features).")

for i in range(N_ITERATIONS):
    # Calculate the loss for this W and B
    W_rand = np.random.randn(d)
    B_rand = np.random.randn()
    current_loss = loss_function(X, y, W_rand, B_rand)
    
    # Check for the best set of parameters found so far
    if current_loss < best_loss:
        best_loss = current_loss
        best_W = W_rand
        best_B = B_rand
        
# --- Final Results ---
print("\n--- Final Random Search Results ---")
print(f"Minimum Loss Found: {best_loss:.4f}")
print("Best Weight Tensor W (first 5 values):")
print(best_W[:5])
print(f"Best Bias Tensor B: {best_B:.4f}")
```
