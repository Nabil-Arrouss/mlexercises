In Naive Bayes, the prior and posterior probabilities are determined by counting the relevant samples from the training set. To perform Naive Bayes classification, these counts are adjusted with a pseudo-counts $\alpha$. When the class probability distribution estimated by MLE (i.e.\ without a prior) is given by 
\begin{align*}
    P(C)=\frac{N_C}{N}
\end{align*}

\begin{align*}
    P(C)=\frac{N_C+\alpha-1}{N+N_{classes}*(\alpha-1)}
\end{align*}
$N_{classes}$ is the number of classes.

### 1. Training
Complete the `nb_train` function. (1 Point)

```python
def nb_train(X, y, alpha=1, K=None, C=None):
    """Train a Naive Bayes model.

    We assume that all features are encoded as integers and have the same domain
    (set of possible values) from 0:(K-1). Similarly, class labels have domain
    0:(C-1).
    """
    N, D = X.shape
    if K is None:
        K = np.max(X) + 1
    if C is None:
        C = np.max(y) + 1

    priors = np.zeros(C)
    for c in range(C):
        priors[c] = np.sum(y == c)
    priors = (priors + alpha) / (N + C * alpha)


    cls = np.zeros((C, D, K))

    for c in range(C):
        Xc = X[y == c]            
        Nc = Xc.shape[0]          

        for j in range(D):
            counts = np.bincount(Xc[:, j], minlength=K)
            cls[c, j, :] = (counts + alpha) / (Nc + K * alpha)


    return dict(logpriors=np.log(priors), logcls=np.log(cls))
```

### 2. Prediction
Complete the `nb_predict` function. (1 Point)

```python
def logsumexp(x):
    """Computes log(sum(exp(x)).

    Uses offset trick to reduce risk of numeric over- or underflow. When x is a
    1D ndarray, computes logsumexp of its entries. When x is a 2D ndarray,
    computes logsumexp of each column.

    Keyword arguments:
    x : a 1D or 2D ndarray
    """
    offset = np.max(x, axis=0)
    return offset + np.log(np.sum(np.exp(x - offset), axis=0))
```

```python
def nb_predict(model, Xnew):
    """Predict using a Naive Bayes model."""
    logpriors = model["logpriors"]   
    logcls = model["logcls"]         
    Nnew = Xnew.shape[0]
    C, D, K = logcls.shape

    logjoint = np.zeros((Nnew, C))

    for c in range(C):
        loglik = np.zeros(Nnew)
        for j in range(D):
            loglik += logcls[c, j, Xnew[:, j]]
        logjoint[:, c] = logpriors[c] + loglik


    yhat = np.argmax(logjoint, axis=1)


    logprob = logjoint[np.arange(Nnew), yhat]

    return dict(yhat=yhat, logprob=logprob)
```

### 4. Model Selection
Complete the code to use cross-validation to find a good value of alpha. (1 Point)

```python
K = 5
Kf = KFold(n_splits=K, shuffle=True)
for i_train, i_test in Kf.split(X):
    print(
        "Fold has {:d} training points and {:d} test points".format(
            len(i_train), len(i_test)
        )
    )
```

```python
def cv_alpha(X, y, alphas, Kfold=5):
    """Return cross-validated accuracy for each alpha."""
    kf = KFold(n_splits=Kfold, shuffle=True, random_state=0)
    accs = []

    for alpha in alphas:
        fold_acc = []
        for train_idx, valid_idx in kf.split(X):
            Xtrain, Xval = X[train_idx], X[valid_idx]
            ytrain, yval = y[train_idx], y[valid_idx]

            model = nb_train(Xtrain, ytrain, alpha=alpha)
            pred = nb_predict(model, Xval)["yhat"]

            fold_acc.append(np.mean(pred == yval))

        accs.append(np.mean(fold_acc))

    return np.array(accs)
```

### 5. Generating Data
Complete the `nb_generate` function to generate new digit images from the learned model. (1 Point)

```python
def nb_generate(model, ygen):
    """Generate data from a trained Naive Bayes model."""
    logcls = model["logcls"]          
    n = len(ygen)
    C, D, K = logcls.shape

    Xgen = np.zeros((n, D), dtype=int)

    for i in range(n):
        c = ygen[i]
        for j in range(D):
            theta = np.exp(logcls[c, j, :])
            Xgen[i, j] = np.random.choice(np.arange(K), p=theta)

    return Xgen
```
